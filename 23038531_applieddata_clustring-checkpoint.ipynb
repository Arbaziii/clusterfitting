```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the dataset
df = pd.read_csv("Mall_Customers.csv")

def relational_graph():
    # Scatter Plot of Annual Income vs. Spending Score
    plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], color='orange')
    plt.title('Annual Income vs. Spending Score')
    plt.xlabel('Annual Income (k$)')
    plt.ylabel('Spending Score (1-100)')
    plt.show()

def categorical_graph():
    # Pie Chart of Gender Distribution
    gender_counts = df['Gender'].value_counts()
    plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title('Gender Distribution')
    plt.show()

def statistical_graph():
    # Histogram of Age Distribution
    plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
    plt.title('Distribution of Customer Ages')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.show()

def line_fitting():
    # Scatter plot of Age vs. Spending Score
    plt.scatter(df['Age'], df['Spending Score (1-100)'], color='orange')

    # Fit a line
    slope, intercept, _, _, _ = linregress(df['Age'], df['Spending Score (1-100)'])
    line = slope * df['Age'] + intercept
    plt.plot(df['Age'], line, color='blue')

    plt.title('Age vs. Spending Score')
    plt.xlabel('Age')
    plt.ylabel('Spending Score (1-100)')
    plt.show()

def kmeans_clustering():
    # Select relevant features for clustering
    X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

    # Fit K-Means clustering algorithm
    kmeans = KMeans(n_clusters=5, random_state=42)
    kmeans.fit(X)

    # Add cluster labels to the dataset
    df['Cluster'] = kmeans.labels_

    # Cluster Visualization
    plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
    plt.title('Customer Segmentation')
    plt.xlabel('Annual Income (k$)')
    plt.ylabel('Spending Score (1-100)')
    plt.legend()
    plt.show()

def elbow_method():
    # Elbow Method
    inertia = []
    for k in range(1, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)

    plt.plot(range(1, 11), inertia, marker='o', color='purple')
    plt.title('Elbow Method')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.show()

def silhouette_analysis():
    silhouette_scores = []
    for k in range(2, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        silhouette_scores.append(silhouette_score(X, kmeans.labels_))

    plt.plot(range(2, 11), silhouette_scores, marker='o', color='green')
    plt.title('Silhouette Analysis')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.show()

# Quality of Clustering
def clustering_quality():
    # Silhouette Score
    kmeans = KMeans(n_clusters=5, random_state=42)
    kmeans.fit(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    print("Silhouette Score:", silhouette_avg)

# Quality of Fitting
def fitting_quality():
    slope, _, _, _, _ = linregress(df['Age'], df['Spending Score (1-100)'])
    print("Slope of the fitted line:", slope)

# Generate and display the plots
relational_graph()
categorical_graph()
statistical_graph()
line_fitting()
kmeans_clustering()
elbow_method()
silhouette_analysis()

# Evaluate clustering and fitting quality
clustering_quality()
fitting_quality()
```
